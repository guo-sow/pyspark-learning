ステップ2: Series の例

Series は pandas の 1 次元ラベル付き array です。データの 1 列を保存し、インデックスを使って個々の値へすばやくアクセスできます。例えば、時間やカテゴリごとの数値データを扱うのに便利です。

- `data`: 実際の値。リスト、配列、辞書などを渡す。
- `index`: 各値のラベル。省略すると 0 から始まる整数ラベルになる。
- `dtype`: データ型を指定できる。省略すると pandas が自動判定する。
- `name`: Series に名前をつけるためのラベル。

第二步: Series 示例

Series 是 pandas 中的一维带标签数组。它可以保存一列数据，并通过索引快速访问每个元素。适合用来表示时间序列、单个字段的数据、或者带索引的列表。

- `data`: 实际的数据值。可以传列表、数组、字典等。
- `index`: 每个值对应的标签。如果省略，默认使用 0 开始的整数索引。
- `dtype`: 指定数据类型。省略时 pandas 会自动推断。
- `name`: 给 Series 起一个名字。

pd.Series([10, 20, 30], index=['a', 'b', 'c'])

```python
s = pd.Series([10, 20, 30], index=['a', 'b', 'c'], name='score')
print(s)
print('index:', s.index)
print('values:', s.values)
print('name:', s.name)
```

ステップ3: DataFrame の説明

DataFrame は pandas の 2 次元ラベル付きデータ構造です。行と列で構成され、表形式のデータを扱うのに適しています。複数の Series をまとめて、異なる列に異なるデータ型を持たせることができます。

第三步: DataFrame 说明

DataFrame 是 pandas 的二维带标签数据结构。它由行和列组成，适合处理表格数据。你可以把多个 Series 组合在一起，每列可以有不同的数据类型。

```python
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie'],
    'score': [90, 85, 95],
    'pass': [True, True, True]
})
print(df)
print('index:', df.index)
print('columns:', df.columns)
```

### リストのリスト作成方式 / 列表的列表创建方式

DataFrame はリストのリストからも作成できます。各内側リストが 1 行分のデータになります。

DataFrame 也可以由列表的列表创建。每个内层列表表示一行数据。


```python
data_list = [
    ['Alice', 90, True],
    ['Bob', 85, True],
    ['Charlie', 95, True]
]
df_list = pd.DataFrame(data_list, columns=['name', 'score', 'pass'])
print(df_list)
```

### NumPy 配列作成方式 / NumPy 数组创建方式

NumPy 配列を使って 2 次元データを DataFrame に変換できます。

可以使用 NumPy 数组来创建 DataFrame。

```python
import numpy as np

rows = 10
cols = 8
arr = np.arange(rows * cols).reshape(rows, cols)

columns = [f'col{i}' for i in range(1, cols + 1)]
index = [f'row{i}' for i in range(1, rows + 1)]

df_array = pd.DataFrame(arr, columns=columns, index=index)

print(df_array)
print('shape:', df_array.shape)
```
